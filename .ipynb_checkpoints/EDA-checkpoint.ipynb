**Cleaning Data**

```python
#checking for null values in the each column of the dataset
data_crash.isna().sum()
```

```python
data_crash.info()
```

```python
#changing datatype of the variables

data_crash['CRASH DATE'] = data_crash['CRASH DATE'].astype('datetime64[ns]')
data_crash['CRASH TIME'] = data_crash['CRASH TIME'].astype('datetime64[ns]')
data_crash['BOROUGH'] = data_crash['BOROUGH'].astype('str')
# data_crash['ZIP CODE'] = data_crash['ZIP CODE'].astype('int')
data_crash['ON STREET NAME'] = data_crash['ON STREET NAME'].astype('str')
data_crash['CROSS STREET NAME'] = data_crash['CROSS STREET NAME'].astype('str')
data_crash['OFF STREET NAME'] = data_crash['OFF STREET NAME'].astype('str')
data_crash['CONTRIBUTING FACTOR VEHICLE 1'] = data_crash['CONTRIBUTING FACTOR VEHICLE 1'].astype('str')
data_crash['CONTRIBUTING FACTOR VEHICLE 2'] = data_crash['CONTRIBUTING FACTOR VEHICLE 2'].astype('str')
data_crash['CONTRIBUTING FACTOR VEHICLE 3'] = data_crash['CONTRIBUTING FACTOR VEHICLE 3'].astype('str')
data_crash['CONTRIBUTING FACTOR VEHICLE 4'] = data_crash['CONTRIBUTING FACTOR VEHICLE 4'].astype('str')
data_crash['CONTRIBUTING FACTOR VEHICLE 5'] = data_crash['CONTRIBUTING FACTOR VEHICLE 5'].astype('str')
data_crash['VEHICLE TYPE CODE 1'] = data_crash['VEHICLE TYPE CODE 1'].astype('str')
data_crash['VEHICLE TYPE CODE 2'] = data_crash['VEHICLE TYPE CODE 2'].astype('str')
data_crash['VEHICLE TYPE CODE 3'] = data_crash['VEHICLE TYPE CODE 3'].astype('str')
data_crash['VEHICLE TYPE CODE 4'] = data_crash['VEHICLE TYPE CODE 4'].astype('str')
data_crash['VEHICLE TYPE CODE 5'] = data_crash['VEHICLE TYPE CODE 5'].astype('str')

#filling null values with 0 and then converting the datatype
data_crash['NUMBER OF PERSONS KILLED'] = data_crash['NUMBER OF PERSONS KILLED'].fillna(0)
data_crash['NUMBER OF PERSONS INJURED'] = data_crash['NUMBER OF PERSONS INJURED'].fillna(0)

data_crash['NUMBER OF PERSONS KILLED'] = data_crash['NUMBER OF PERSONS KILLED'].astype('int')
data_crash['NUMBER OF PERSONS INJURED'] = data_crash['NUMBER OF PERSONS INJURED'].astype('int')

print("Datatype conversion completed!")
```
